```python
import pandas as pd
```

```python
df = pd.read_csv('titanic_dataset.csv')
df.head(10)
```

```python
dependent_var = df.drop(['PassengerId','SibSp','Name','Ticket','Parch','Cabin','Embarked','Survived'],axis='columns')
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dependent_var.Sex = dependent_var.Sex.map({'male': 1, 'female': 2})
dependent_var.Age = dependent_var.Age.fillna(dependent_var.Age.mean())
dependent_var
```

```python
independent_var = df['Survived']
independent_var
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(dependent_var,independent_var,test_size = 0.3)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
